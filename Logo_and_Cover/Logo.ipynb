# Logo

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from math import log10
```

Choose data to shape each rings of donut chart

```python
labels = list("ABCDEFGHIJK")
data = [1,2,3,4,5,6,7,8,9,9.9999999,9.9999999]
```

Number of data points

```python
n = len(data)
```

Find max value for full ring

```python
k = 10 ** int(log10(max(data)))
m = k * (1 + max(data) // k)
```

Radius of donut chart

```python
r = 2
```

Calculate width of each ring

```python
w = r / n 
```

Create colors along a chosen colormap

```python
colors = [cm.viridis(i / n) for i in range(n)]
```

Create rings of donut chart

```python
fig, ax = plt.subplots()
ax.axis("equal")

for i in range(n):
    innerring, _ = ax.pie([m - data[i], data[i]], radius = r - i * w, startangle = 90, labels = ["", labels[i]], labeldistance = 1 - 1 / (1.5 * (n - i)), textprops = {"alpha": 0}, colors = ["white", colors[i]])
    plt.setp(innerring, width = w, edgecolor = "none")
    
plt.show()
```

Save the final logo 

```python
plt.savefig('DataShow_Logo.png', transparent=True, dpi=1200, bbox_inches='tight')
```
